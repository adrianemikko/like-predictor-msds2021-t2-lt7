## Data discussion

### About the Dataset
The dataset includes 14,090 instances of Facebook statuses including 9 features. The data was gathered using each of one of the authors' group member’s Facebook tokens and included data from that individual and all of their friends. The features that were collected include the number of friends, age, and gender of the user who posted the status, the time of the status (month and times of day), the time since the last status (hours), and the “score” of the Facebook status itself. To find the score of a Facebook status, a dictionary was built by the authors with individual scores of keywords from statuses in their entire dataset. The individual score was calculated by averaging the number of likes that a status with that word receives. Once the individual scores for each word is calculated and stored in a dictionary, the status is scored by averaging those values for each word in the status. 

An additional column was added to bin the number of likes *(**0** for 0 to 10 likes, **1** for 11 to 100 likes, and **2** for 101 to 1000 likes)* and that was the feature that the group wanted to predict, which is included in the training and validation set. The training set included 75% of the total data instances and was used to train our model while the testing set contained 3,522 instances used to evaluate the model’s results.

### Features

The dataset has `nine (9)` features as shown below: 

- **`friend_count:`** This number could help indicate the range of values that a user’s status could receive, since this would be roughly the maximum number of likes.
- **`age:`** The user's age. The user’s group of friends may consist mostly of other user’s of the same age, and people of this age may have certain interests.
- **`gender:`** Statuses about certain topics may be viewed differently if posted by one gender.
- **`Time the status was posted (month and time of day):`** Posting statuses about time-sensitive events will garner more likes if posted in a timely manner. Additionally, posting statuses when more people are online will also increase visibility.
    - **`month:`** Month when the post was posted on their facebook profile/page.  
    - **`time_of_day:`**
        - post_midnight: 12AM - 4AM
        - early_morning: 4AM - 8AM
        - morning: 8AM - 11AM
        - noon: 11AM - 2PM
        - afternoon: 2PM - 5PM
        - evening: 5PM - 9PM
        - night: 9PM - 12AM
- **`time_since_last (hours):`** People may be less inclined to like the status of someone who spams.
- **`message_score`** - A score assigned based on the words used in the status.
- **`average_user_likes:`** If a person get a high number of likes on every status, their statuses will get like more in general.
- **`score:`** Score of the status based off of the word score dictionary. Certain words and topics can contribute more likes to a status than others. To calculate the score, we remove common words, such as “the”, “was”, “am”, etc, and assign each of the remaining words a score of (the natural log of the number of likes in this status) / (number of remaining words). The score for each word is updated and averaged in the word score dictionary. We take the natural log so that statuses that have the same number of likes within an order of magnitude will be classified similarly.

##  kNN implementation

### Import Libraries

```python
# import mglearn #library provided by amueller
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import gridspec as gridspec
from mpl_toolkits.axes_grid1 import make_axes_locatable
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

%matplotlib inline
```

### Define Functions

```python
def plot_acc_vs_k(X, y, kmax=50, tprop=0.25, nloops=100):
    # Declare tables and arrays
    data = []

    for k in range(1, kmax + 1, 1):
        for n in range(nloops):
            # Split the data frame
            __ = (train_test_split(X, y, test_size=tprop, random_state=n))
            X_train, X_test, y_train, y_test = __

            # Build the model
            clf = KNeighborsClassifier(n_neighbors=k)
            clf.fit(X_train, y_train)

            # Record training and test accuracy
            data.extend([[k, n, 0, clf.score(X_train, y_train)],
                         [k, n, 1, clf.score(X_test, y_test)]])

    # Frame data, get the optimal k, and corresponding accuracy
    df = pd.DataFrame(data, columns=['k', 'n', 'set', 'accuracy'])
    acc_diff = (df.query('set == 0').groupby('k')['accuracy'].mean()
                - df.query('set == 1').groupby('k')['accuracy'].mean())
    d2 = acc_diff.abs().to_frame()
    d2['Test Accuracy'] = df.query('set == 1').groupby('k')['accuracy'].mean()
    d2.reset_index(inplace=True)

    # Plot the data
    fig = plt.figure(figsize=(15, 15), constrained_layout=True)
    gs = gridspec.GridSpec(6, 2, figure=fig)

    ax = fig.add_subplot(gs[0:3, :])
    sns.lineplot(x='k',
                 y='accuracy',
                 data=df,
                 hue='set',
                 estimator='mean',
                 ci='sd',  # Use standard deviation as `confidence interval`
                 legend='full',
                 ax=ax)
    ax.set_title('Accuracy vs Number of Nearest Neighbors',
                 fontsize=25, fontweight='black') # 16
    ax.legend(['Training Accuracy', 'Test Accuracy'], loc=0)
    ax.set_ylabel('$Accuracy$', fontsize=16)
    ax.set_xlabel('$n$ $Neighbors$', fontsize=16)
    ax.grid()
    ax.set_yticks(np.arange(0, 1.1, 0.1))
    ax.set_xticks(range(0, kmax + 1, kmax//10))
    ax.set_ylim(0, 1.1)  # Set limits to 0-100% accuracy
    ax.set_xlim(1, kmax)
    make_axes_locatable(ax).append_axes('right', 
                                        size='5%',
                                        pad=0.1).axis('off')
    
    ax2 = fig.add_subplot(gs[3:5, :])
    cax = make_axes_locatable(ax2).append_axes('right', size='5%', pad=0.1)
    g = ax2.scatter(d2['k'],
                    d2['accuracy'],
                    c=d2['Test Accuracy'],
                    s=1250,
                    cmap='viridis_r')
    cbar = fig.colorbar(g, cax=cax, orientation='vertical')
    cbar.set_label('$Accuracy$')
    ax2.set_title('Accuracy Difference vs Number of Nearest Neighbors',
                  fontsize=16, fontweight='black')
    ax2.set_ylabel('$Accuracy$ $Difference$')
    ax2.set_xlabel('$n$ $Neighbors$')
    ax2.grid()
    ax2.set_xticks(range(0, kmax + 1, kmax//10))
    ax2.set_xlim(1, kmax)

    # Return results
    return ax, df
```

## Discussion of results

75% data was used for training the model, and 25% was reserved for testing. To build the model, we used k-Nearest Neighbors (kNN) learning algorithms to test various combination of features. Predicting popularity based on the number of likes (<= 3 as unpopular and >3 as popular) using kNN, we were able to build a model with **62.32% accuracy** at **13 neighbors**, which was the optimal number of nearest neighbors.

To improve the accuracy, we tried to divide the number of likes into bins or buckets.  The buckets were 0 - 10 likes, 11 - 100 likes, and 100+ likes. By categorizing the number of likes into bins, we were able to better predict the number of likes a certain individual's post will have in this distinct target variable. An accuracy of **84.6%** was obtained with the optimal number of nearest neighbors at **39**. To further explore the dataset, we also tried to predict other features using kNN algorithm and one of those features is gender. The reason why gender was used because there's a "theory" that women gets more likes in Facebook compared to men. We tested this theory by using some of the features from our dataset to see if the kNN can accurately classify the gender of the user who posted on Facebook. The result shows that it could accurately predict the gender with at least **80% accuracy** as long as the number of nearest neighbors is **below 65.** The behavior of the graph for this experiment is quite different compared to the other experiments. Rather than there's a plateau, the graph goes downward as the number of nearest neighbors increases. The lowest accuracy difference is usually where the accuracy is also the lowest.

Overall, user-defined features drive the number of likes per post, which suggest that building repertoire and popularity come hand-in-hand. Truly, popularity is an iterative process, involving building ones' features little by little, post-by-post.
